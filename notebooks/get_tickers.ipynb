```python

url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object from the response content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table containing the S&P 500 data
table = soup.find('table', {'id': 'constituents'})

# Extract the rows from the table
rows = table.find_all('tr')
stocks = []
# Skip the header row and iterate over the remaining rows
for row in rows[1:]:
    # Extract the columns from the row
    columns = row.find_all('td')
    # Extract the company name and section from the columns
    id = columns[0].text.strip()
    company_name = columns[1].text.strip()
    sector = columns[2].text.strip()
    stocks.append({"id": id, "company_name": company_name, "sector": sector})
    # Print the company name and section
    print(f"ID: {id}")
```
